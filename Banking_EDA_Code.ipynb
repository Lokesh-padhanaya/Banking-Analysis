```python
bins=[0, 100000, 300000, float('inf')]
labels = ['Low', 'Med', 'High']

df['Income Band'] = pd.cut(df['Estimated Income'], bins = bins, labels = labels, right = False)
```

```python
df['Income Band'].value_counts()
```

```python
# Examine the distribution of unique categories in categorical columns
categorical_cols = df[['Nationality', "Fee Structure",'Amount of Credit Cards','Occupation','Loyalty Classification','Properties Owned','Risk Weighting','BRId','GenderId','IAId','Income Band']].columns

for col in categorical_cols:
    print(f"\nValue counts for '{col}':")
    display(df[col].value_counts())
```

## Univariate Analysis

```python
for i, predictor in enumerate(df[['Nationality', "Fee Structure",'Amount of Credit Cards','Occupation','Loyalty Classification','Properties Owned','Risk Weighting','BRId','GenderId','IAId','Income Band']]):
    plt.figure(i)
    sns.countplot(data=df, x= predictor)
    
```

## Bivariate Analysis

```python
for i, predictor in enumerate(df[['Nationality', "Fee Structure",'Amount of Credit Cards','Occupation','Loyalty Classification','Properties Owned','Risk Weighting','BRId','GenderId','IAId','Income Band']]):
    plt.figure(i)
    sns.countplot(data=df, x= predictor, hue = 'GenderId')
```

```python
for i, predictor in enumerate(df[['Nationality', "Fee Structure",'Amount of Credit Cards','Occupation','Loyalty Classification','Properties Owned','Risk Weighting','BRId','GenderId','IAId','Income Band']]):
    plt.figure(i)
    sns.countplot(data=df, x= predictor, hue = 'Nationality')
```

```python
# Histplot of value counts for different Occupation

for col in categorical_cols:
    if col == "Occupation":
        continue
    plt.figure(figsize=(8,4))
    sns.histplot(df[col])
    plt.title('Histogram of Occupation count')
    plt.xlabel(col)
    plt.ylabel("count")
    plt.show()
```

## Numerical Analysis

```python
numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']

# Univariate Analysis and Visualization 
plt.figure(figsize=(16,8))
for i, col in enumerate(numerical_cols):
    plt.subplot(3,3,i+1)
    sns.histplot(df[col],kde=True)
    plt.title(col)   
plt.tight_layout() 
plt.show()
```

## Heatmaps

```python
numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']
correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(10,10))
sns.heatmap(correlation_matrix,annot=True,cmap='crest',fmt='.2f')
plt.title("Correlation Matrix")
plt.title("Correlation Matrix")
plt.show()
```

## Insights of EDA:

1. The strongest positive correlation occur among "Bank Deposits" with "Checking Accounts", "Saving Accounts" and "Foreign Currency Accounts" idicating that customers who maintain high balance in one account type often hold substantial amount/funds across other accounts as well.
